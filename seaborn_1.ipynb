```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('salary_data.csv')
df.head()
```

```python
sns.scatterplot(x='Age',y='Salary', data=df)
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='Age',y='Salary', data=df)
```

```python
sns.scatterplot(x='Age',y='Salary', data=df, hue = 'Years of Experience' )
```

```python
sns.scatterplot(x='Age',y='Salary', data=df, hue = 'Salary' )
```

```python
sns.scatterplot(x='Age',y='Salary', data=df, hue = 'Years of Experience', palette='dark' )
```

```python
sns.scatterplot(x='Age',y='Salary', data=df, hue = 'Years of Experience', palette='Dark2' )
```

```python
sns.scatterplot(x='Age',y='Salary', data=df, hue = 'Years of Experience',alpha=0.2)
```

```python
sns.regplot(x='Age',y='Salary',data=df)
```

```python
plt.figure(figsize=(12,5),dpi=700)
sns.scatterplot(x='Age',y='Salary', data=df, hue = 'Salary',alpha=0.2)
plt.show()
```
