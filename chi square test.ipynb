```python
observed_values = data.values
print(f"observed values is {observed_values}")
```

```python
chi2_stats,p_value,dof,expected_values = stats.chi2_contingency(data)
```

```python
p_value
```

```python
expected_values
```

```python
chi2_stats
```

```python
alpha = 0.05
critical_values = stats.chi2.ppf(q = 1-alpha, df = dof)
print(f"critical value is {critical_values}")
```

```python
if chi2_stats <= critical_values:
    print('Accept H0')
    
else:
    print("Reject H0")
```

```python
if p_value <= alpha:
    print('Reject H0')
    
else:
    print("Accept H0")
```
