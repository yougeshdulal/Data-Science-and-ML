```python
import numpy as np
import matplotlib.pyplot as plt
arr1 = np.array([1,3,5,7,9,11])
arr2 = np.array([22,55,33,77,1001,11])
plt.plot(arr1,arr2,marker='*')
plt.savefig('Demo1')
plt.show()
```

```python
arr1 = np.array([1,3,5,7,9,11])
arr2 = np.array([22,55,33,77,1001,11])
plt.plot(arr1,arr2,marker='D')
plt.savefig('Demo1')
plt.show()
```

```python
arr1 = np.array([1,3,5,7,9,11])
arr2 = np.array([22,55,33,77,1001,11])
plt.plot(arr1,arr2,marker='d')
plt.savefig('Demo1')
plt.show()
```

```python
arr1 = np.array([1,3,5,7,9,11])
arr2 = np.array([22,55,33,77,1001,11])
plt.plot(arr1,arr2,marker='d', mfc='red')
# mfc marker face color
plt.savefig('Demo1')
plt.show()
```

```python
arr1 = np.array([1,3,5,7,9,11])
arr2 = np.array([22,55,33,77,1001,11])
plt.plot(arr1,arr2,marker='d', mfc='red', mec ='black')
# mfc marker face color
# mec marker edge color
plt.savefig('Demo1')
plt.show()
```

```python
arr1 = np.array([1,3,5,7,9,11])
arr2 = np.array([22,55,33,77,1001,11])
plt.plot(arr1,arr2,marker='d', mfc='red', mec ='black')
# mfc marker face color
# mec marker edge color
plt.xlabel("Array1---->")
plt.ylabel("Array2---->")
plt.savefig('Demo1')
plt.show()
```

```python
arr1 = np.array([1,3,5,7,9,11])
arr2 = np.array([22,55,33,77,1001,0])
plt.plot(arr1,arr2,marker='d', mfc='red', mec ='black')
# mfc marker face color
# mec marker edge color
plt.title("Demo of matplotlib")
plt.xlabel("Array1---->")
plt.ylabel("Array2---->")
plt.savefig('Demo1')
plt.show()
```
