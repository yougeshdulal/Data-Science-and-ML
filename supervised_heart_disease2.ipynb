```python
pd.crosstab(df.cp,df.target).plot(kind='bar',figsize=(8,6),color=['crimson','lightblue'])
# lets communicate more now
plt.title('Heart disease freq per chest pain')
plt.xlabel('chest pain type')
plt.ylabel('amount')
plt.legend(['disease','no disease'])
```
