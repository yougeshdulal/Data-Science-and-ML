```python
import matplotlib.pyplot as plt
%matplotlib inline
a1 = np.array([22,33,44,55])
color=['red','green','blue','lightblue']
plt.pie(a1,colors=color)
plt.show()
```

```python
a1 = np.array([22,33,44,55])
color=['red','green','blue','lightblue']
labels=['first','second','third','fourth']
plt.pie(a1,colors=color,labels=labels)
plt.show()
```

```python
a1 = np.array([22,33,44,55])
color=['red','green','blue','lightblue']
explode = (0,0.1,0,0)
labels=['first','second','third','fourth']
plt.pie(a1,colors=color,labels=labels,explode=explode,
       shadow=True,counterclock=False,
        labeldistance=1.1,radius=1.3
       )
plt.show()
```
