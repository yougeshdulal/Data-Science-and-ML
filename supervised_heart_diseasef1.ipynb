```python
# split data into X and y
X = df.drop('target',axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split

# target ma duita cha vanesi binary classification ko kura ho
# aba data lai training and test ma divide garum
# splitting data into training and test sets
np.random.seed(42)  
# X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7)
```

```python
y_train,len(y_train)
```

```python
# train test data set vayo aba machine learning model banaune 
# train garne (finding patterns) on the training set
# test garne (using patterns) on the test set
# chosing the ml models
#  we will be using 3 different ML models  
# scikit learn
# Logistic regression
# k nearest Neighbours classifier
# random forest classifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# putting models in dictionary
models={
    'Logistic Regression': LogisticRegression(),
    'KNN':KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier()
}
```

```python
# create a function to fit and score models
def fitandscoremodels(models,X_train,X_test,y_train,y_test):
    """
    fits and evaluates ML models
    models: dict of scikit learn ml models
    X_train: no labels(training data)
    X_test: testing data(no labels)
    y_train: training labels
    y_test: testing labels
    """
#     setting random seeds so that results are reproducable
    np.random.seed(42)
#     make a dict to keep model scores
    model_scores={}
#     looping through models
    for name,model in models.items():
#         fit the model to the data
        model.fit(X_train,y_train)
#     evaluate the model and add the score to the model_scores
        model_scores[name]=model.score(X_test,y_test)
    return model_scores
    
```

```python
model_scores=fitandscoremodels(models=models,X_train=X_train,X_test=X_test,
                              y_train=y_train,
                              y_test=y_test)
model_scores
```

```python
# comparing the models
```

```python
model_compare = pd.DataFrame(model_scores,index=['accuracy'])
```
