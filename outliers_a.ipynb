```python
# z score
outliers = []
# function to take data as argument and return possible outliers
def detect_outliers(data):
    threshold = 3 # 3 std detaviation
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
detect_outliers(dataset)
```

```python
dataset = [10,11,13,25,27,31,34,55,61,91,35,91,77,81,56,65,76,67,53,54,500,5000]
```

```python
detect_outliers(dataset)
sns.histplot(dataset,kde=True)
```
